```python
import datasets
from datasets import load_dataset, concatenate_datasets
import numpy as np

# Many of the 21 corpora have sizes: 10K, 30K, 100K, 300K, 1M but not all of them
# All except Slovak was available for 2019
# Change the ids below to try different sizes or years
data_ids = [
    "bul_news_2019_10K",
    "ces_news_2019_10K",
    "dan_news_2019_10K",
    "deu_news_2019_10K",
    "ell_news_2019_10K",
    "eng_news_2019_10K",
    "est_news_2019_10K",
    "fin_news_2019_10K",
    "fra_news_2019_10K",
    "hun_news_2019_10K",
    "ita_news_2019_10K",
    "lav_news_2019_10K",
    "lit_news_2019_10K",
    "nld_news_2019_10K",
    "pol_news_2019_10K",
    "por_news_2019_10K",
    "ron_news_2019_10K",
    "slk_news_2020_10K",
    "slv_news_2019_10K",
    "spa_news_2019_10K",
    "swe_news_2019_10K"
]

language_map = {
    "bul": "Bulgarian",
    "ces": "Czech",
    "dan": "Danish",
    "deu": "German",
    "ell": "Greek",
    "eng": "English",
    "est": "Estonian",
    "fin": "Finnish",
    "fra": "French",
    "hun": "Hungarian",
    "ita": "Italian",
    "lav": "Latvian",
    "lit": "Lithuanian",
    "nld": "Dutch",
    "pol": "Polish",
    "por": "Portuguese",
    "ron": "Romanian",
    "slk": "Slovak",
    "slv": "Slovenian",
    "spa": "Spanish",
    "swe": "Swedish"
}

def get_multi_language_dataset(data_ids):
    
    datasets = []
    
    for data_id in data_ids:
        ds = load_dataset("imvladikon/leipzig_corpora_collection", data_id, split="train", trust_remote_code=True)
        language_code = data_id.split('_')[0]
        label = language_map[language_code]
        ds = ds.add_column("label", [label] * len(ds))
        datasets.append(ds)

    full_dataset = concatenate_datasets(datasets)

    return full_dataset

full_ds = get_multi_language_dataset(data_ids)

# dataset information
print("Dataset info: ", full_ds)
print("Dataset shape: ", full_ds.shape)
print("Labels: ", set(full_ds["label"]))
print("Sample data: ", full_ds["sentence"][:1], full_ds["label"][:1])
```

```python
import regex as re

## Preprocessing the data and sentences to remove punctiouation and put text to lowercase
word_pattern = re.compile(r"\p{L}+", flags=re.UNICODE) ## use of regex because of caracters such as romanian caracters

clean_texts = []
clean_labels = []

for item in full_ds:
    label = item["label"]
    text = item["sentence"]

    words = word_pattern.findall(text.lower()) ## find all words corresponding to the pattern
    cleaned = " ".join(words)

    clean_texts.append(cleaned)
    clean_labels.append(label)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(clean_texts, clean_labels,test_size=0.2, random_state=42, stratify=clean_labels)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,test_size=0.3, random_state=42, stratify=y_train)

## Checking the sizees of the train and test splits
print("Training set size: ", len(X_train))
print("Validation set size: ", len(X_val))
print("Test set size: ", len(X_test))
print("Training labels distribution: ", {label: y_train.count(label) for label in set(y_train)})

X_val[:5], y_val[:5]  # show some sample data
```

```python
from pathlib import Path
import regex as re

## Map language codes to full language names
language_map_test = {
    "en": "English",
    "fr": "French",
    "et": "Estonian",
    "sv": "Swedish",
    "ro": "Romanian",
    "it": "Italian",
    "da": "Danish"
}

languages_path_list = list(Path("./Test_dataset").rglob("*"))
selected_languages = {}
for path in languages_path_list:
    if path.is_dir():
        continue
    lang = path.suffix.lstrip(".")
    if lang in language_map_test and lang not in selected_languages:
        selected_languages[lang] = path

print("Selected files:", selected_languages)

word_pattern = re.compile(r"\p{L}+", flags=re.UNICODE)

X_test_final = []
y_test_final = []

N_LINES = 5000    ## limit number of lines per language for testing for computation cost 

for lang_code, file_path in selected_languages.items():

    lang_name = language_map_test[lang_code]
    count = 0

    with open(file_path, "r", encoding="utf-8", errors="ignore") as f:

        for line in f:
            ## extract words with fitting pattern
            words = word_pattern.findall(line.lower())
            if len(words) == 0:
                continue

            # reconstitute cleaned text
            cleaned_text = " ".join(words)

            X_test_final.append(cleaned_text)
            y_test_final.append(lang_name)

            count += 1
            if count >= N_LINES:
                break

print("Test samples:", len(X_test_final))
print("Example:", X_test_final[0])

X_test_final[:5], y_test_final[:5]  # show some sample data
```

### Task 1.2

```python
import numpy as np
from collections import Counter
from sklearn.metrics.pairwise import cosine_similarity

def rotate(vec, shift):
    return np.roll(vec, shift)

def bind(vectors):
    v = vectors[0]
    for x in vectors[1:]:
        v = v * x
    return v


def build_hd_centroids(X_train, y_train, d=100):
    ## Collect all characters in training data
    all_text = "".join(X_train)
    chars = np.unique(list(all_text))
    encoding_chars = {c: np.random.choice([-1,1], size=d) for c in chars} ## generate random hypervectors for each character

    ## Merge texts per language
    lang_texts = {lang: "" for lang in set(y_train)}
    for txt, lang in zip(X_train, y_train):
        lang_texts[lang] += txt

    ## Build centroids with ngrams thanks to the ArticleEncodingNgramStatistics.pdf method
    centroids = {}
    for lang, txt in lang_texts.items():
        grams = [tuple(txt[i:i+3]) for i in range(len(txt)-2)]
        counts = Counter(grams) ## count occurrences of each trigram with Counter because the python version was taking for ever

        S = np.zeros(d)
        for gram, cnt in counts.items():
            g1 = encoding_chars[gram[0]]
            g2 = rotate(encoding_chars[gram[1]], 1)
            g3 = rotate(encoding_chars[gram[2]], 2)
            S += cnt * bind([g1, g2, g3])

        S /= np.linalg.norm(S) ## L2 normalization
        centroids[lang] = S

    return centroids, encoding_chars

## Encode text function used for testing on smaller text 
def encode_text(text, encoding_chars, d):
    grams = [tuple(text[i:i+3]) for i in range(len(text)-2)]
    S = np.zeros(d)
    for g in grams:
        for code in g:
            if code not in encoding_chars:
                encoding_chars[code] = np.random.choice([-1,1], size=d) ## added because encountered unknown characters in test set
        g1 = encoding_chars[g[0]]
        g2 = rotate(encoding_chars[g[1]], 1)
        g3 = rotate(encoding_chars[g[2]], 2)
        S += bind([g1, g2, g3])
    norm = np.linalg.norm(S)
    if norm > 0:
        S /= norm ## added because otherwise the similarity calculation failed with division by zero
    else:
        S = np.random.choice([-1,1], size=d)
    return S

def predict(text, centroids, encoding_chars):
    encoded_text = encode_text(text, encoding_chars, d=len(next(iter(centroids.values())))) 
    sims = {lang: cosine_similarity(encoded_text.reshape(1,-1), centroid.reshape(1,-1))[0][0] for lang,centroid in centroids.items()}
    return max(sims, key=sims.get) ## extract the language with the highest similarity

```

Question : what will be the size of the n-gram input vector in conventional (local)
representation?

For n grams, the one hot vector has the dimension corresponding to every possible n gram meaning :
$$
size = a^n
$$
with $a$ = alphabet size and $n$ = length of n gram


```python
n = 3  ## n-gram size

languages_text = {lang: "" for lang in set(y_train)}
for text, label in zip(X_train, y_train):
    languages_text[label] += text  

## Compute alphabet size per language
alphabet_sizes = {}
for lang, text in languages_text.items():
    unique_chars = np.unique(list(text)) ## get all unique characters in the text
    alphabet_sizes[lang] = len(unique_chars) ## store the size of the alphabet

print(f"Alphabet sizes per language: {alphabet_sizes}")
print("\n")

## Compute n-gram vector size for each language
print("n-gram vector sizes:")
for lang, a in alphabet_sizes.items():
    print(f"{lang}: {a**n}")
```

As we can notice, the dimensionality explodes for certain languages making the feature vectores huge and therefore the computational and memory management complicated. One hot encoding also doesn't allow compositionaly meaning combining features or binding structures. It also isn't robust against noise. 

### Task 1.3

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score
from sklearn.metrics.pairwise import cosine_similarity 
import matplotlib.pyplot as plt
import seaborn as sns
dimensions = [100,1000]

for d in dimensions:
    centroids, encoding_chars = build_hd_centroids(X_train, y_train, d=d)

    y_pred = [predict(x, centroids, encoding_chars) for x in X_val]
    print(f"Dimension {d}, F1:", f1_score(y_val, y_pred, average="macro")) ## get f1 score with predictions from the validation set
    print(f"Dimension {d}, Accuracy:", accuracy_score(y_val, y_pred))

    cm = confusion_matrix(y_val, y_pred, labels=sorted(centroids.keys())) ## plot confusion matrix

    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=sorted(centroids.keys()), yticklabels=sorted(centroids.keys()))
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix")
    plt.tight_layout()
    plt.show()
```

```python
dimensions = [100, 1000]

target_languages = list(language_map_test.values())

## Filter training set to only include target languages
mask = [label in target_languages for label in y_train]
X_train_small = [x for x, m in zip(X_train, mask) if m]
y_train_small = [y for y, m in zip(y_train, mask) if m]

for d in dimensions:
    centroids, encoding_chars = build_hd_centroids(X_train_small, y_train_small, d=d)

    y_pred = [predict(x, centroids, encoding_chars) for x in X_test_final]

    print(f"Dimension {d}, F1:", f1_score(y_test_final, y_pred, average="macro"))
    print(f"Dimension {d}, Accuracy:", accuracy_score(y_test_final, y_pred))

    cm = confusion_matrix(y_test_final, y_pred, labels=sorted(centroids.keys()))
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=sorted(centroids.keys()),
                yticklabels=sorted(centroids.keys()))
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(f"Confusion Matrix (d={d})")
    plt.tight_layout()
    plt.show()
```
